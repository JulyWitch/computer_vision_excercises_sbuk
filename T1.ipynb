## Zooming 

We will zoom an image using opencv method and then we will do it manually pixel by pixel



### Using cv2.resize

First step is importing libraries

```
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

Now we have to load our image and print the quantity of pixels

```
img = cv2.imread('lenna.png')
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

As you see we our image has 512 * 512 pixels 

So in this step we will add new pixels only in width of the image so you can see new pixels clearly

```
height, width = img.shape[:2]

resized_img = cv2.resize(img, (2*width,height), interpolation = cv2.INTER_NEAREST)
print("INTER_NEAREST method")
plt.imshow(resized_img)
```

And also for both width and height

```
resized_img = cv2.resize(img, (2*width,2*height), interpolation = cv2.INTER_NEAREST)
print("INTER_NEAREST method")
plt.imshow(resized_img)
```

## Shrinking 


### Using cv2.resize
So we just need to give new height and width to this function and then it will set other things and gives us a cool image


```
import cv2
import matplotlib.pyplot as plt 
import numpy as np

img = cv2.imread('lenna.png')
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)

height, width = img.shape[:2]

resized_img = cv2.resize(img, ( int(width * 0.5), int(height * 0.5)) )
print("INTER_AREA method")
plt.imshow(resized_img)
```

## Convert an image to negative










```
import cv2
import matplotlib.pyplot as plt 
import numpy as np

img = cv2.imread('lenna.png')
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
height, width = img.shape[:2]

for row in range(height):
    for col in range(width):
        red = 255 -img[row][col][0] 
        green = 255 -img[row][col][1] 
        blue = 255 -img[row][col][2] 
        img[row][col]=[red,green,blue]

plt.imshow(img)
```

## Use log and exp operators on an image

```
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
   
image = cv2.imread('lenna.png') 
img = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)

c = 255 / np.log(1 + np.max(image)) 
log_image = c * (np.log(image + 1)) 

log_image = np.array(log_image, dtype = np.uint8) 
   
plt.imshow(img) 
plt.show() 
plt.imshow(log_image) 
plt.show() 
```

# Excercise 5


```
import matplotlib.pyplot as plt
import cv2

img = cv2.imread('cameraman.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
plt.show()
flat = img.flatten()

# show the histogram
plt.hist(flat, bins=50)
plt.show()

# create our own histogram function
def get_histogram(image, bins):
    # array with size of bins, set to zeros
    histogram = np.zeros(bins)
    
    # loop through pixels and sum up counts of pixels
    for pixel in img:
        histogram[pixel] += 1
    
    # return our final result
    return histogram

# execute our histogram function
hist = get_histogram(flat, 256)


def reHist(myHist):
    myHist = iter(myHist)
    newHist = [next(myHist)]
    for i in myHist:
      newHist.append(newHist[-1]+i)
    return np.array(newHist)

histEq = reHist(hist)

# display the result
plt.plot(histEq)
plt.show()


img_new = histEq[flat]

# put array back into original shape since we flattened it
img_new = np.reshape(img_new, img.shape)

plt.imshow(img_new, cmap='gray')

plt.show(block=True)

```
